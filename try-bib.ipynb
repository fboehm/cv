```python
import citeproc
from citeproc.source.json import CiteProcJSON
from citeproc_styles import get_style_filepath
from yaml import load, Loader

# Load CSL file
with open('csl_files/chicago-author-date-16th-edition.csl', 'r') as csl_file:
    csl = csl_file.read()

# Load YAML file with references
with open('data/boehm.yaml', 'r') as yaml_file:
    references = load(yaml_file, Loader=Loader)

# Initialize citation style
style = citeproc.CitationStylesStyle(get_style_filepath('apa'))

# Create citation processor
citation_style = citeproc.CitationStylesBibliography(style, citeproc.source.json.CiteProcJSON(CiteProcJSON(csl)))

# Sort references in reverse chronological order
references.sort(key=lambda x: x['issued']['date-parts'][0], reverse=True)

# Add references to the citation processor
for reference in references:
    citation = citeproc.Citation([citeproc.CitationItem(reference['id'])])
    citation_style.register(citation)

# Print the formatted bibliography
bibliography = citation_style.bibliography()
for entry in bibliography:
    print(entry.render(style))
```
